```python
# Import the pathlib and csv library
from contextlib import nullcontext
from pathlib import Path
import csv

# Set the file paths
menupath = Path('../PyRamen/Resources/menu_data.csv')
salespath = Path('../PyRamen/Resources/sales_data.csv')


# Initialize Dictionary
report = {}
sale_item = {}


# Open the sales csv file as an object 
with open(salespath, 'r') as salefile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the salereader object
    salereader = csv.reader(salefile, delimiter=',')

    # Read the header row first (skip this step if there is no header)
    sale_header = next(salereader) 
    # Read each row of data after the header
    for row in salereader:

        # Set the 'name', 'count', and 'revenue' variables for better
        # readability, convert strings to ints for numerical calculations
        num_items = row[0]
        sale_date = row[1]
        sale_credit = row[2]
        sale_quan = int(row[3])
        menu_item = row[4]
        
        #Check if menu item existed in report dictionary, add sale item if not. 
        #Update 01-count total qunanity for each ramen type
        if menu_item not in report:
            sale_item = {"01-count":0,"02-revenue":0, "03-cogs":0,"04-profit":0}
            sale_item["01-count"] = sale_quan
            report[menu_item] = sale_item
            
        elif menu_item in report:
            report[menu_item]["01-count"] += sale_quan
      

    # Open the menu csv file as an object
with open(menupath, 'r') as menufile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(menufile, delimiter=',')

    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader) 


    # Read each row of data after the header
    for row in csvreader:

        # Set the 'items', 'category', 'descr', 'price', and 'cost' variables for better
        # readability, convert strings to ints for numerical calculations
        items = row[0]
        category = row[1]
        descr = row[2]
        price = float(row[3])
        cost = int(row[4])
        quantity = 0
        profit = price-cost

        #Check if menu item existed in report dictionary, add sale item if not. 
        if items in report:
            quantity = report[items]["01-count"]
            report[items]["02-revenue"] += price * quantity
            report[items]["03-cogs"] += cost * quantity
            report[items]["04-profit"] += profit * quantity
        else:
            print(f" ", items, "does not equal to items in Report! NO MATCH!")
            
    # Print Report Dictionary
    print(report)
   

# Set the path for the output.txt
output = Path("../PyRamen/output.txt")

# Open the output path as a file and pass into the 'csv.writer()' function
# Set the delimiter/separater as a ','
with open(output, 'w') as txtfile:
    txtfile.writelines("Ichiban Ramen Financial Report by menu item \n")
    txtfile.writelines("--------------------------------------------------------\n\n")
    #loop through write the dictionary to txt file
    for x, y in report.items():
        txtfile.writelines(f'{x}{y}\n\n')

    txtfile.close    
```
